### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
results_hygeine20 = establishments.find(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"There are {count} establishments with a hygeine score of 20")


# Display the first document in the results using pprint
pprint(results_hygeine20[0])
```

```python
import pandas as pd

# Convert the result to a Pandas DataFrame
df_hygeine20 = pd.DataFrame(results_hygeine20)

# Display the number of rows in the DataFrame
print(len(df_hygeine20))

# Display the first 10 rows of the DataFrame
df_hygeine20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and RatingValue greater than or equal to 4
query = {"AddressLine3": "London", "RatingValue": {"$gte": 4}}
results_london4 = establishments.find(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"There are {count} establishments in London with a rating of 4 or higher")

# Display the first document in the results using pprint
pprint(results_london4[0])
```

```python
# Convert the result to a Pandas DataFrame
df_london4 = pd.DataFrame(results_london4)

# Display the number of rows in the DataFrame
print(len(df_london4))

# Display the first 10 rows of the DataFrame
df_london4.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find_one({'BusinessName': 'Penang Flavours'})['geocode']['latitude']
longitude = establishments.find_one({'BusinessName': 'Penang Flavours'})['geocode']['longitude']

query = {
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
    'RatingValue': 5
}
sort_by =  {'scores.Hygiene': -1}

# store
results_degree = establishments.find(query).sort(list(sort_by.items()))

# Print the results
results_list = list(establishments.find(query).sort(sort_by))

for result in results_list:
    print(result)
```

```python
# Convert result to Pandas DataFrame
df_degree = pd.DataFrame(results_degree)
df_degree.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Assuming 'establishments' is your MongoDB collection

# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

query_p4 = {"scores.Hygiene": 0}
group_by_p4 = {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}
sort_by_p4 = {"$sort": {"count": -1}}

pipeline_p4 = [
    {"$match": query_p4},
    group_by_p4,
    sort_by_p4
]

# Use the aggregate method to execute the pipeline
results_p4 = list(establishments.aggregate(pipeline_p4))

# Print the number of documents in the result
count_p4 = establishments.count_documents(query_p4)
print(f"The number of docs: {count_p4}")

# Print the results
for result in results_p4[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame
df_p4 = pd.DataFrame(results_p4)

# Display the number of rows in the DataFrame
print(len(df_p4))

# Display the first 10 rows of the DataFrame
df_p4.head(10)
```
